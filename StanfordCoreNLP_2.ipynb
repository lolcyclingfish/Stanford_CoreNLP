```python
nlp = StanfordCoreNLP('http://localhost:9000')
```

## 1 - Tokenizatoin in StanfordCoreNLP

```python
text_file1=open('wsj_0063.txt', 'r',encoding="utf8")
text_file1=text_file1.read()
print(text_file1)
```

```python
text_split1 = nlp.annotate(text_file1, properties={
  'annotators': 'ssplit, tokenize',
  'outputFormat': 'json'
  })
```

```python
#text_split1
```

```python
#join split tokens into sentences:
sentences=[]
for s in text_split1['sentences']:
    sentences.append(' '.join([t['word'] for t in s['tokens']]))
sentences
```

## 2 - Normalization in StanfordCoreNLP

```python
text_norm1 = nlp.annotate(text_file1, properties={
  'annotators': 'tokenize, lemma',
  'outputFormat': 'json'
  })
```

```python
#text_norm1
```

```python
#join split lemmas into sentences:
sentences1=[]
for s in text_norm1['sentences']:
    sentences1.append(' '.join([t['lemma'] for t in s['tokens']]))
sentences1
```

## Q3 - Tokenization & Normalization in StanfordCoreNLP

```python
#read in classbio2.txt as list of sentences

text_file2=open('classbios.txt', 'r',encoding="utf8")
text_file2=text_file2.read()
#print(text_file2)
```

```python
#tokenize and normalize
text_split2 = nlp.annotate(text_file2, properties={
  'annotators': 'tokenize, lemma',
  'outputFormat': 'json'
  })
```

```python
#text_split2
```

```python
#join split lemmas into sentences:
sentences2=[]
for s in text_split2['sentences']:
    sentences2.append(' '.join([t['lemma'] for t in s['tokens']]))
#sentences2
```

## 4. Basic frequency analysis

Write a program to calculate the frequency of each (normalized) word in the normalized classbios.txt corpus. Report the top 20 most frequent words, excluding stop words and punctuation, and their frequencies. Describe whether you think this list gives much information about this corpus. Did you learn anything from this list about the backgrounds of the class?

```python
#split each normalized sentences into words in sentences2, and append into one list:
list_of_words=[]
for i in range(len(sentences2)):
    list_of_words=list_of_words+sentences2[i].split()

#list_of_words
```

```python
import nltk
#nltk.download()
```

```python
#remove stop words using nltk
from nltk.corpus import stopwords

reduced_list_of_words=[]
stopset = ['I','a','and','to','the','in','of','my','for','with','that','as','at','from','is','on','have','me','be','an','it','this',
          ',', '.', '------------------------------------------------------------', ':', '**']
cachedStopWords = stopset+stopwords.words("english")

reduced_list_of_words=reduced_list_of_words+[word for word in list_of_words if word not in cachedStopWords]
#reduced_list_of_words
```

```python
#count frequency of words in the list and return the top 10 most frequent words:
import collections
counts=collections.Counter(reduced_list_of_words)
counts.most_common(20)
```
